```python
# Creación y representación de los grafos G y H.
import networkx as nx
import matplotlib.pyplot as plt

# Creación del grafo G.
G = nx.Graph()
G.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F', 'G'])
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'C'), ('D', 'E'), ('E', 'F'), ('E', 'G')])

# Creación del grafo H.
H = nx.Graph()
H.add_node('H')
H.add_nodes_from(G)
H.add_edges_from(G.edges)

# Representación de los grafos G y H.
plt.figure(figsize=[8.0,4.0], dpi=200)
plt.subplot(121,title='Grafo G')
nx.draw(G, with_labels=True, node_color = 'c', font_color='r', node_size=140)
plt.subplot(122, title='Grafo H')
nx.draw(H, with_labels=True, node_color = 'b', font_color='w', node_size=140)
```

__Un grafo completo de n vértices o nodos, tiene n*(n-1)/2 aristas, y se denota Kn.__

A continuación se muestra el número de aristas para K1, K2, K3, K4, K5, K6 Y K99.

```python
import networkx as nx
#Creación de los grafos completos K1, K2, K3, K4, K5, K6 y K99.
G01=nx.complete_graph(1)
G02=nx.complete_graph(2)
G03=nx.complete_graph(3)
G04=nx.complete_graph(4)
G05=nx.complete_graph(5)
G06=nx.complete_graph(6)
G99=nx.complete_graph(99)
n=1
print('Número de aristas de  K1= {:>5}'.format(G01.number_of_edges()), 'Fórmula  n*(n-1)/2 ={:>5}'.format(int(n*(n-1)/2)))
n=2
print('Número de aristas de  K2= {:>5}'.format(G02.number_of_edges()), 'Fórmula  n*(n-1)/2 ={:>5}'.format(int(n*(n-1)/2)))
n=3
print('Número de aristas de  K3= {:>5}'.format(G03.number_of_edges()), 'Fórmula  n*(n-1)/2 ={:>5}'.format(int(n*(n-1)/2)))
n=4
print('Número de aristas de  K4= {:>5}'.format(G04.number_of_edges()), 'Fórmula  n*(n-1)/2 ={:>5}'.format(int(n*(n-1)/2)))
n=5
print('Número de aristas de  K5= {:>5}'.format(G05.number_of_edges()), 'Fórmula  n*(n-1)/2 ={:>5}'.format(int(n*(n-1)/2)))
n=6
print('Número de aristas de  K6= {:>5}'.format(G06.number_of_edges()), 'Fórmula  n*(n-1)/2 ={:>5}'.format(int(n*(n-1)/2)))
n=99
print('Número de aristas de K99= {:>5}'.format(G99.number_of_edges()), 'Fórmula  n*(n-1)/2 ={:>5}'.format(int(n*(n-1)/2)))
```
