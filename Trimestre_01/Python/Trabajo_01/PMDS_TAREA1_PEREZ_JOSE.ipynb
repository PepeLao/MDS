## Problema 1
Escribir una función que leyendo un número entero $n$ sea capaz de imprimir un pozo de Mario Bros de $n+1$ escalones cuya separación sea de $n$ espacios.

```python
#----------------------------------
# Ingrese su solución en esta celda
#----------------------------------
def PozoMarioBross(n=3):
    s=''

    for i in range(n+1):
        s+=' '*(n-i)+'#'*(i+1)+' '*(n)+'#'*(i+1)+'\n'
    return s

# Ingrese el valor entero n, en la función PozoMarioBross(). Por defecto n=3.
print(PozoMarioBross(5))
```

```python
# Se importa la librería math para usar las funciones math.pi y math.sin.
# Se define la función rad, que será usada para convertir ángulos a radianes.
import math
rad = lambda a:(math.pi/180)*a

#-----------------------------------------------------
# Ingrese su solución para la función alcance_maximo R
#-----------------------------------------------------
# Función R(), con parámetros por defecto a = 45 grados, v = 100 m/s y g = 9.81 m/(s*s)
def R(a=45, v=100, g=9.81):
    
    r=(v*v*math.sin(2*rad(a)))/g
#Formateo de salida
    salida  = ''
    salida += 'Con ' + str(a) + ' grados, '
    salida += 'a una velocidad de ' + str(v) + ' m/s, '
    salida += 'y con una aceleración de gravedad de ' + str(g) + ' m/s^2'
    salida += ', el alcance máximo R es ' + str(r) + ' m.'
    return salida
# Si se quiere solo retorna el valor r, comentar formateo de salida y descomentar linea siguiente
#    return r

#----------------------------------------------------
# Ingrese su solución para la función altura_maxima H
#----------------------------------------------------
# Función H(), con parámetros por defecto a = 45 grados, v = 100 m/s y g = 9.81 m/(s*s)
def H(a=45, v=100, g=9.81):
    
    h=(v*v*math.sin(rad(a)))/(2*g)
#Formateo de salida
    salida  = ''
    salida += 'Con ' + str(a) + ' grados, '
    salida += 'a una velocidad de ' + str(v) + ' m/s, '
    salida += 'y con una aceleración de gravedad de ' + str(g) + ' m/s^2'
    salida += ', la altura máxima H es ' + str(h) + ' m.'
    return salida
# Si se quiere solo retorna el valor h, comentar formateo de salida y descomentar linea siguiente
#    return h

#---------------------------------------------------------------------------------------
# Solución para la función alcance_maximo y altura_maxima en una misma función Proyectil
#---------------------------------------------------------------------------------------
# Función Proyectil(), con parámetros por defecto a = 45 grados, v = 100 m/s y g = 9.81 m/(s*s)
def Proyectil(a=45, v=100, g=9.81):

    r=(v*v*math.sin(2*rad(a)))/g
    h=(v*v*math.sin(rad(a)))/(2*g)
#Formateo de salida
    salida  = ''
    salida += 'Con ' + str(a) + ' grados, '
    salida += 'a una velocidad de ' + str(v) + ' m/s, '
    salida += 'y con una aceleración de gravedad de ' + str(g) + ' m/s^2, se tiene: \n'
    salida += 'El alcance máximo R es ' + str(r) + ' m.\n'
    salida += 'La altura máxima H es ' + str(h) + ' m.\n'
    return salida    
# Si se quiere retornar valores r y h, comentar formateo de salida y descomentar linea siguiente    
#    return r, h


# Ingrese angulo a y velocidad v para obtener alcance máximo R.
# Por defecto a = 45 grados, v = 100 m/s y g = 9.81 m/(s*s)
print(R())
# Ingrese angulo a y velocidad v para obtener altura máxima H.
# Por defecto a = 45 grados, v = 100 m/s y g = 9.81 m/(s*s)
print(H())
# Ingrese angulo a y velocidad v para obtener alcance máximo R y altura máxima H.
# Por defecto a = 45 grados, v = 100 m/s y g = 9.81 m/(s*s)
print(Proyectil())
```

La cantidad de piezas $f(n)$ logradas con $n$ cortes está dada por la siguiente recurrencia:

$$ f(n) = n+f(n-1)$$

Implemente una función recursiva en `python` que reciba el número de cortes y devuelva el máximo número de piezas posibles.

```python
#----------------------------------
# Ingrese su solución en esta celda
#----------------------------------
def CortadorPerezoso(n=0):
    if n==0:
        return 1
    else:
        return n + CortadorPerezoso(n-1)
    
# Ingrese el valor de n un valor entero, para la función CortadorPerezoso. Por defecto n=0.
CortadorPerezoso()
```

## Problema 4
El algoritmo babilónico para encontrar una aproximación de la raíz cuadrada $g$ de un número $x$ viene dada por los siguientes pasos:
1. Partir con un número de adivinanza $g$
2. Si $g\times g$ está suficientemente cercano a $x$, entonces $g$ es una buena aproximación de la raíz de $x$.
3. Si no, crear una nueva estimación promediando $g$ y $x/g$, es decir $$g'=\dfrac{1}{2}\left(g+\dfrac{x}{g}\right) $$
4. Usando $g'$, volver al paso 2.

Implemente este algoritmo dentro de una función que reciba como argumento el número al que se le desea calcular la raíz cuadrada, y que tenga como argumento por defecto una variable `epsilon` con la _tolerancia_ de la aproximación dada por `epsilon=1e-3`. 

```python
#----------------------------------
# Ingrese su solución en esta celda
#----------------------------------
# Solución 1, usando iteraciones, con un máximo de 100 iteraciones.
# Por defecto x = 9, y epsilon = 1.0e-3. Adivinanza g es igual a x.
def raiz2i(x=9,epsilon=1.0e-3):
    g=x
    iteracion=100
    for i in range(iteracion):
        if abs(x-g**2)<epsilon:
            return g
        else:
            g=(1/2*(g+x/g))
            

# Solución 2, usando recursividad.
# Por defecto x = 9, adivinanza g = 10 y epsilon = 1.0e-3.
def raiz2r(x=9, g=10, epsilon=1.0e-3):
      
    if abs(x-g**2)<epsilon:
        return g
    else:
        return raiz2r(x, (1/2)*(g+x/g))

# Ingrese el valor x, para conocer su raíz cuadrada usando iteraciones.
# Puede también ingresar el valor epsilon.
# Por defecto x = 9, y epsilon = 1.0e-3
print(raiz2i())

# Ingrese el valor x, para conocer su raíz cuadrada usando recursividad.
# Puede también ingresar el valor adivinanza g, y el valor epsilon.
# Por defecto x = 9, g = 10 y epsilon = 1.0e-3.
print(raiz2r())
```

Con esta información, implemente una función en `python` que acepte un mensaje (string) y lo encripte usando una llave definida por defecto. Para esto defina una variable global `alfabeto="abcdefghijklmnñopqrstuvwxyz"`.

```python
#----------------------------------
# Ingrese su solución en esta celda
#----------------------------------
# Se define función CesarCifra(). Por defecto el mensaje es 'hola mundo¡' y key = 3.
def CesarCifra(s='hola mundo¡',key=3):
    alfabeto='abcdefghijklmnñopqrstuvwxyz'
    la=len(alfabeto)
    ls=len(s)
    mensaje=''
    
    for i in range(ls):
        x=s[i:i+1]
        for j in range(la):
            if s[i:i+1]==alfabeto[j:j+1]:
                x=alfabeto[(j+key)%la:(j+key)%la+1]
        mensaje+=x
        
    return mensaje


# Se define función CesarDescifra(). Por defecto el mensaje es 'krñd oxpgr¡' y key = 3.
def CesarDescifra(s='krñd oxpgr¡',key=3):
    alfabeto='abcdefghijklmnñopqrstuvwxyz'
    la=len(alfabeto)
    ls=len(s)
    mensaje=''
    
    for i in range(ls):
        x=s[i:i+1]
        for j in range(la):
            if s[i:i+1]==alfabeto[j:j+1]:
                x=alfabeto[(j-key)%la:(j-key)%la+1]
        mensaje+=x            

    return mensaje



# Ingrese el mensaje s que quiere cifrar y llave key que quiere usar.
# Por defecto s = 'hola mundo¡' y key=3
print(CesarCifra())

# Ingrese el mensaje s que quiere descifrar y llave key que quiere usar.
# Por defecto s = 'krñd oxpgr¡',key=3
print(CesarDescifra())

# PRUEBAS
# Para probar el cifrado y descifrado use CesarDescifra(CesarCifra())
# Use un mensaje que tenga las letras de inicio y de fin de alfabeto, como condición de borde.
#CesarCifra('hola amigo, puedes leer este mensaje?')
#CesarDescifra('krñd oxpgr¡ def abc')
print(CesarDescifra(CesarCifra('Hola mundo¡ abc xyz')))
```
