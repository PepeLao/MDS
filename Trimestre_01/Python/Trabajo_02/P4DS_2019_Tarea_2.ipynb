## Problema 1

Escribir un módulo `bitacora.py` que sea capaz de manejar un archivo de texto como una bitácora. Este módulo debe tener las siguientes funciones:

- `abrir_bitacora(<nombre_archivo>)`: Función que abre un archivo de texto de nombre `<nombre_archivo>`. Si no existe, lo crea.
- `guardar_bitacora(<nombre_archivo>, mensaje)`: Función que agrega la entrada `mensaje` (string) en el archivo `<nombre_archivo>`. 

La entrada debe tener el siguiente formato:
`[ 2019:07:30 21:43 ] mensaje entrada`, es decir `[ AÑO:MES:DIA HORA:MIN ] MENSAJE`.

- `cerrar_bitacora(<nombre_archivo>, mensaje=“Fin del registro”)`: Función que cierra el archivo agregando `mensaje` (string con un valor por defecto) al final.

Investigue y use los módulos `os` y `datetime` para su implementación.

Adjunte adicionalmente el módulo como un script con nombre `apellido_inicial-nombre_bitacora.py` en el formulario.

```python
#---------------------------------------------------
# Ingrese su solución para bitácora.py en esta celda
#---------------------------------------------------
# Módulo bitacora.py para manejar un archivo de texto como una bitácora
# Importa módulo os.
import os
#Importa módulo datetime.
import datetime as dt

def abrir_bitacora(nombre_archivo='bitacora.txt'):
    """Obtiene el Path del archivo. Sirve para Windows, Mac, Linux."""
    ruta_archivo=os.path.join(os.getcwd(),nombre_archivo)
    
    """Si el archivo existe:
         Abre el archivo en modo agregar data al final del archivo.
    """    
    if os.access(ruta_archivo, os.F_OK):
        puntero_archivo=open(ruta_archivo,'a')
    else:
        """Si el archivo no existe, lo crea y lo abre para escritura."""
        puntero_archivo=open(ruta_archivo,'w')
        
    """Retorna el puntero al archivo."""
    return puntero_archivo
        
def guardar_bitacora(puntero_archivo, mensaje='Hola mundo'):
    """Formatea la fecha y hora del mensaje"""
    fecha_hora = dt.datetime.now().strftime('[ %Y:%m:%d %H:%M ] ')
    """Formatea el registro a ser ingresado en el archivo."""
    registro=fecha_hora + mensaje + '\n'
    """Escribe el registro en el archivo """
    puntero_archivo.write(registro)


def cerrar_bitacora(puntero_archivo, mensaje='Fin del registro'):
    """Formatea la fecha y hora del mensaje"""
    fecha_hora = dt.datetime.now().strftime('[ %Y:%m:%d %H:%M ] ')
    """Formatea el registro a ser ingresado en el archivo."""
    registro=fecha_hora + mensaje + '\n'
    """Escribe el registro en el archivo """
    puntero_archivo.write(registro)
    """Cierra el archivo """
    puntero_archivo.close()

def mostrar_bitacora(nombre_archivo='bitacora.txt'):
    
    """Obtiene el Path del archivo. Sirve para Windows, Mac, Linux."""
    ruta_archivo=os.path.join(os.getcwd(),nombre_archivo)
    
    with open(ruta_archivo) as ptro:
        datos_leidos = ptro.read()
    
    return datos_leidos
    
ptr_arc=abrir_bitacora('prueba.txt')
guardar_bitacora(ptr_arc,'Mensaje de prueba2')
cerrar_bitacora(ptr_arc)    
```

1. Una función que lea las primeras `n` líneas.

```python
#-----------------------------
# Ingrese su solución parte 1
#-----------------------------
# Se importa el módulo os
import os
# Funcíón Lee_primeras_n_lineas()
# Argumentos
#   <nom_arc> : nombre del archivo, por defecto 'Dracula_chapter_I.txt'
#   n : número n de primeras líneas del archivo, por defecto 10.
def Lee_primeras_n_lineas(nom_arc='Dracula_chapter_I.txt',n=10):

    """Obtiene el Path del archivo. Sirve para Windows, Mac, Linux."""
    ruta_archivo=os.path.join(os.getcwd(),nom_arc)

    """Declara el string s."""
    s=''
    
    """Si el archivo existe:
         Abre el archivo en modo lectura.
         Lee las primeras n líneas del archivo y las guarda en s.
         Cierra el archivo.
    """
    if os.access(ruta_archivo, os.F_OK):
        puntero_archivo=open(ruta_archivo,'r')

        for i in range(n):
            s+=puntero_archivo.readline()

        puntero_archivo.close()
        
    else:
        """Si el archivo no existe, escribe un mensaje en s."""
        s='Archivo no encontrado. Asegurese que este en la ruta :' + ruta_archivo

    """Retorna s"""
    return s

print(Lee_primeras_n_lineas('Dracula_chapter_I.txt',15))
```

2. Una función que imprima el texto entre las líneas `n` y `m` (argumentos de la función).

```python
# Se importa el módulo os
import os
# Funcíón Imprime_entre_n_y_m_lineas()
# Argumentos
#   <nom_arc> : nombre del archivo, por defecto 'Dracula_chapter_I.txt'
#   n : número n de inicio, por defecto 1.
#   m : número m de finicio, por defecto 15.
def Imprime_entre_n_y_m_lineas(nom_arc='Dracula_chapter_I.txt',n=5,m=15):
    """Obtiene el Path del archivo. Sirve para Windows, Mac, Linux."""
    ruta_archivo=os.path.join(os.getcwd(),nom_arc)

    """Declara el string s."""
    s=''

    """Si el archivo existe:
         Abre el archivo en modo lectura.
         Lee las primeras n-1 líneas del archivo sin guardalas.
         Lee las siguientes m-n del archivo y las guarda en s.
         Cierra el archivo.
    """
    if os.access(ruta_archivo, os.F_OK):
        puntero_archivo=open(ruta_archivo,'r')
        for i in range(m):
            if i>=n-1:
                s+=puntero_archivo.readline()
            else:
                puntero_archivo.readline()
                
        puntero_archivo.close()
    else:
        """Si el archivo no existe, escribe un mensaje en s."""
        s='Archivo no encontrado. Asegurese que este en la ruta :' + ruta_archivo

    """Retorna s"""
    return s

print(Imprime_entre_n_y_m_lineas('Dracula_chapter_I.txt',1, 3))
```

4. Una función que encuentre la palabra más larga de la línea `n` (dada por el usuario).

```python
#-----------------------------
# Ingrese su solución parte 3
#-----------------------------
# Se importa el módulo os
import os

# Función blxce, reemplaza los caracteres especiales por blanco.
blxce = lambda s : s.translate({ord(c): ' '  for c in '"!@#$%^&*()[]{};:,./<>?\|`~-=_+'})

# Función Palabra_mas_larga
# Argumentos:
#   <nom_arc> nombre de archivo, por defecto 'Dracula_chapter_I.txt'
#   lin es el número de línea del archivo a seleccionar, y obtener la palabra + larga.
def Palabra_mas_larga(nom_arc='Dracula_chapter_I.txt',lin=13):

    """Obtiene el Path del archivo. Sirve para Windows, Mac, Linux."""
    ruta_archivo=os.path.join(os.getcwd(),'Dracula_chapter_I.txt')

    """Declara el la lista lis."""
    lis=[]

    """Si el archivo existe:
         Abre el archivo en modo lectura.
         Con un ciclo for se posiciona en linea buscada y la guarda en s.
         Cierra el archivo.
         Reemplaza caracteres especiales por blanco en s. Splitea s y lo guarda en lis.
         Obtiene la palabra más larga de lis.
         Escribe mensaje en s, indicando palabra más larga.
    """    
    if os.access(ruta_archivo, os.F_OK):
        puntero_archivo=open(ruta_archivo,'r')
        
        for i in range(lin):
            s=puntero_archivo.readline()
            
        puntero_archivo.close()
        
        lis=blxce(s).split()
 
        maslarga=' '
        for palabra in lis:
            if len(palabra)>len(maslarga):
                maslarga=palabra
                
        s='La palabra más larga de la línea ' + str(lin) + ' es ' + maslarga
    else:
        """Si el archivo no existe, escribe un mensaje en s."""
        s='Archivo no encontrado. Asegurese que este en la ruta :' + ruta_archivo

    """Retorna s"""
    return s

print(Palabra_mas_larga('Dracula_chapter_I.txt',55))
```

```python
#-----------------------------
# Ingrese su solución parte 4
#-----------------------------

preps = ['about', 'above', 'across', 'after', 'against', 'along', 'among', 'around', 'as', 'at', 'before',
         'behind', 'below', 'beneath', 'beside', 'between', 'beyond', 'by', 'despite', 'down', 'during', 
         'except', 'for', 'from', 'in', 'inside', 'into', 'like', 'near', 'of', 'off', 'on', 'onto', 'opposite',
         'out', 'outside', 'over', 'past', 'round', 'since', 'than', 'through', 'to', 'towards', 'under', 
         'underneath', 'unlike', 'until', 'up', 'upon', 'via', 'with', 'within', 'without' ]

pronouns = ['I', 'you', 'he', 'she', 'it', 'we', 'they']

# Se importa el módulo os
import os
# Función blxce, reemplaza los caracteres especiales por blanco.
blxce = lambda s : s.translate({ord(c): ' '  for c in '"!@#$%^&*()[]{};:,./<>?\|`~-=_+'})


def Frecuencia_de_palabras(lista=pronouns):
    """
       Obtiene el Path del archivo. Sirve para Windows, Mac, Linux
    """
    ruta_archivo=os.path.join(os.getcwd(),'Dracula_chapter_I.txt')
    
    """
       Declaración de lista lis.
       Declaración de diccionario dic.
    """
    lis=[]
    dic={}

    """
       Abre archivo y lo almacena en datos_leidos.
       Todas las palabras en datos_leidos, se dejan en minúsculas.
    """
    with open(ruta_archivo) as ptro:
        datos_leidos = ptro.read().lower()

    """
       Reemplaza carácteres especiales por blancos en datos_leidos.
       Splitea datos_leidos y almacena en lista lis.
    """    
    lis=blxce(datos_leidos).split()

    """
       Almacena en diccionario dic las palabras de la lista de entrada de la función.
       Cuenta las ocurrencias de las palabras en lis y guarda en dic.
    """
    for elemento in lista:
        dic[elemento]=lis.count(elemento.lower())

    """
       Retorna el diccionario dic.
    """
    return dic

#print(Frecuencia_de_palabras(preps))
print(Frecuencia_de_palabras(pronouns))
```
